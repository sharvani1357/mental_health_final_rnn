```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
sentence=["I","Love","Deep","Learning"]
print(sentence)
word_embeddings={
    "I":1.0,
    "Love":2.0,
    "Deep":3.0,
    "Learning":4.0
}
print(word_embeddings)
x=np.array([
    [
    [word_embeddings["I"]],
    [word_embeddings["Love"]],
    [word_embeddings["Deep"]],
    [word_embeddings["Learning"]]
    ]
])
print(x)
```

```python
y=np.array([word_embeddings['Learning']])
print(y)
print(x.shape)
print(y.shape)
```

```python
#Building simple rnn
model=Sequential()
model.add(SimpleRNN(units=8,activation='tanh',input_shape=(3,1)))
model.add(Dense(1))
```

```python
#compiling the model of rnn
model.compile(optimizer='adam',loss='mse')
```

```python
#training the model
model.fit(x,y,epochs=300,verbose=1)
```

```python
#predicting the output
test_seq=np.array([
    [1.0],[2.0],[3.0]
])
pred=model.predict(test_seq)
print("prediction:",pred[0][0])
```

```python
#convert the predicted word into the close word
predicted_num=pred[0][0]
close_word=min(word_embeddings,key=lambda word:abs(word_embeddings[word]-predicted_num))
print("predicted word:",close_word)
```
